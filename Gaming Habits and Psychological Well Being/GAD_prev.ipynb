```python
import pandas as pd


df["Playstyle"] = df["Playstyle"].str.lower().str.strip()

# Correct typos and inconsistencies
df["Playstyle"] = df["Playstyle"].str.replace("stragers", "strangers")


def categorize_playstyle(playstyle):
    if "singleplayer" in playstyle:
        return "Singleplayer"
    elif "multiplayer" in playstyle and "friends" in playstyle:
        return "Multiplayer with Friends"
    elif "multiplayer" in playstyle and "strangers" in playstyle:
        return "Multiplayer with Strangers"
    elif "multiplayer" in playstyle and "online" in playstyle:
        return "Multiplayer - Online"
    elif "multiplayer" in playstyle and ("real life" in playstyle or "offline" in playstyle):
        return "Multiplayer - Offline"
    elif "multiplayer" in playstyle:
        return "Multiplayer-Unknown"
    else:
        return "Unknown"

# Apply the function to create the "Playstyle_category" column
df["Playstyle_category"] = df["Playstyle"].apply(categorize_playstyle)

# 3. Replace Playstyle with Category (optional)

# Uncomment the following line if you want to replace the original "Playstyle" column
df["Playstyle"] = df["Playstyle_category"]

# 4. Drop Playstyle_category column (if replaced)

# Uncomment the following line if you replaced the original "Playstyle" column
data = df.drop(columns=["Playstyle_category"])

# 5. Print the first few rows to verify the changes
df.head()
```

```python
import pandas as pd
import plotly.express as px
from scipy import stats

# Assuming 'data' is your cleaned pandas DataFrame

# 1. Group Data by Work Status
grouped_data = data.groupby("Work")

# 2. Analyze Hours Played by Work Status

# ANOVA test
f_statistic, p_value = stats.f_oneway(*[group["Hours"] for name, group in grouped_data])
print("ANOVA for Hours Played:")
print(f"F-statistic: {f_statistic:.2f}, p-value: {p_value:.4f}")

# Interactive bar chart with Plotly Express
fig = px.bar(grouped_data["Hours"].mean().reset_index(), x="Work", y="Hours", 
             title="Average Hours Played by Work Status",
             labels={"Hours": "Average Hours Played"})
fig.show()

# 3. Analyze Playstyle Preferences by Work Status

# Chi-square test
playstyle_crosstab = pd.crosstab(data["Work"], data["Playstyle"])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(playstyle_crosstab)
print("\nChi-square test for Playstyle Preferences:")
print(f"Chi-square statistic: {chi2_stat:.2f}, p-value: {p_value:.4f}")

# Interactive stacked bar chart with Plotly Express
fig = px.bar(playstyle_crosstab.stack().reset_index(), x="Work", y=0, color="Playstyle",
             title="Playstyle Preferences by Work Status",
             labels={0: "Proportion of Players"})
fig.show()

# 4. Analyze GAD_T Scores by Work Status

# Kruskal-Wallis H-test (non-parametric alternative to ANOVA)
h_statistic, p_value = stats.kruskal(*[group["GAD_T"] for name, group in grouped_data])
print("\nKruskal-Wallis H-test for GAD_T Scores:")
print(f"H-statistic: {h_statistic:.2f}, p-value: {p_value:.4f}")

# Interactive box plot with Plotly Express
fig = px.box(data, x="Work", y="GAD_T", 
             title="GAD_T Scores by Work Status",
             labels={"GAD_T": "GAD_T Score"})
fig.show()
```
