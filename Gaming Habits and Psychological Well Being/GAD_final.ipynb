```python
import pandas as pd
import plotly.express as px
from scipy.stats import chi2_contingency

# Define age groups
def get_age_group(age):
    if age >= 18 and age <= 20:
        return '18-20'
    elif age >= 21 and age <= 25:
        return '21-25'
    elif age >= 26 and age <= 34:
        return '26-34'
    elif age >= 45 :
        return '45+'

# Group the data by 'whyplay_category' and 'age_group' and calculate counts
df['age_group'] = df['Age'].apply(get_age_group)
reason_age_group = df.groupby(['whyplay_category_2', 'age_group']).size().reset_index(name='Count')

# Calculate total counts for each age group
total_counts = reason_age_group.groupby('age_group')['Count'].sum().reset_index(name='Total')

# Merge total counts with reason_age_group
reason_age_group = reason_age_group.merge(total_counts, on='age_group')

# Calculate proportion of each reason for playing games within each age group
reason_age_group['Proportion'] = reason_age_group['Count'] / reason_age_group['Total']

# Create a grouped bar plot
fig = px.bar(reason_age_group, x='age_group', y='Proportion', color='whyplay_category_2',
             title='Reasons for Playing Games by Age Group',
             labels={'age_group': 'Age Group', 'Proportion': 'Proportion of Participants',
                     'whyplay_category_2': 'Reason for Playing Games'})

# Chi-square test
observed = pd.crosstab(df['whyplay_category_2'], df['age_group'])
chi2, p, dof, expected = chi2_contingency(observed)
print("Chi-square statistic:", chi2)
print("p-value:", p)

# Display the grouped bar plot
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf

data = df


from scipy import stats

def kruskal_effect_size(h_statistic, n, k):
  """
  Calculates epsilon-squared and eta-squared effect sizes for Kruskal-Wallis test.

  Args:
      h_statistic: The H-statistic from the Kruskal-Wallis test.
      n: The total sample size.
      k: The number of groups.

  Returns:
      A tuple containing epsilon-squared and eta-squared effect sizes.
  """
  epsilon_squared = h_statistic / (n**2 - 1)
  eta_squared = (h_statistic - k + 1) / (n - 1)
  return epsilon_squared, eta_squared


# Perform Shapiro-Wilk test
shapiro_stat, shapiro_p_value = stats.shapiro(data["SWL_T"])

# Print the statistical test results
print("Shapiro-Wilk test statistic:", shapiro_stat)
print("p-value:", shapiro_p_value)






r, p_value = stats.spearmanr(data["Hours"], data["GAD_T"])
print(f"Correlation between Hours and GAD_T: r = {r:.4f}, p-value = {p_value:.4f}")


h_statistic, p_value = stats.kruskal(*[group["Hours"] for name, group in data.groupby("SPIN_T")])
print(f"Kruskal-Wallis for Hours Played by SPIN_T: H-statistic = {h_statistic:.2f}, p-value = {p_value:.4f}")


epsilon_squared, eta_squared = kruskal_effect_size(h_statistic, len(data), len(data["SPIN_T"].unique()))
print(f"Eta-squared effect size for SPIN_T: {eta_squared:.4f}")




sns.jointplot(x="Hours", y="SPIN_T", data=data, kind="reg", marginal_kws=dict(kde=True))
plt.suptitle("Hours Played vs. SPIN_T with Marginal Distributions", y=1.02)
plt.show()

sns.jointplot(x="Hours", y="GAD_T", data=data, kind="reg", marginal_kws=dict(kde=True))
plt.suptitle("Hours Played vs. GAD_T with Marginal Distributions", y=1.02)
plt.show()

sns.jointplot(x="Hours", y="SWL_T", data=data, kind="reg", marginal_kws=dict(kde=True))
plt.suptitle("Hours Played vs. SWL_T with Marginal Distributions", y=1.02)
plt.show()


```

However, the effect size (eta-squared = 0.0116) is small, suggesting that playstyle explains only a small portion of the variance in hours played.
